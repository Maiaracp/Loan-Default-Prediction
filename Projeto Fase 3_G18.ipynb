```python
#Separar valores data

def encode_dates(df, column):
    df = df.copy()
    df[column] = pd.to_datetime(df[column], format='%d-%m-%y')
    df[column + '_year'] = df[column].apply(lambda x: x.year)
    df[column + '_month'] = df[column].apply(lambda x: x.month)
    df[column + '_day'] = df[column].apply(lambda x: x.day)
    df = df.drop(column, axis=1)
    return df

```

### ---Regressão Logística---

```python
X = treino.drop(['loan_default'], axis=1)
y = treino['loan_default']
```

```python
#função calcular acurácia Kfold
def intervalo(results):
    mean = results.mean()
    dv = results.std()
    print('Acurácia média: {:.2f}%'.format(mean*100))
    print('Intervalo de acurácia: [{:.2f}% ~ {:.2f}%]'
           .format((mean - 2*dv)*100, (mean + 2*dv)*100))
```

```python
#particionando base teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, stratify = y)
```

```python
## Padronização das variáveis

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index, columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index, columns=X_test.columns)
```

```python
X_train.head()
```

```python
#treinar o modelo e calcular acurácia
cv = StratifiedKFold(n_splits = 5, shuffle = True)
model = LogisticRegression(max_iter = 1000)
results = cross_val_score(model, X_train, y_train, cv = cv)
intervalo(results)
```

### Comparação de Modelos

```python
#Função para treinamendo de modelos
acuracia = {}
roc_r = {}

def treina_modelo(modelo, nome):
    print(nome)
    
    model = modelo.fit(X_train_bl, y_train_bl)
    pred = modelo.predict(X_test_bl)
    
    #Métricas de comparação de modelos
    acc = accuracy_score(y_test_bl, pred)*100
    acuracia[nome] = acc
    print('\nAcurácia:',acc)
    print('\nPrecisão:',precision_score(y_test_bl, pred)*100)
    print('\nRecall:',recall_score(y_test_bl, pred)*100)
    print('\nf1_score:',f1_score(y_test_bl, pred)*100)
    
    
    #ROC
    roc = roc_auc_score(y_test_bl, pred)*100
    roc_r[nome] = roc
    print('\nroc_auc',roc)
    
    # Matriz de Confusão
    print('\nMatriz de Confusão')
    print(pd.DataFrame(confusion_matrix(y_test_bl, pred)))
    
    #ROC Score
    fpr, tpr, threshold = roc_curve(y_test_bl, pred)
    roc_auc = auc(fpr, tpr)*100
    
    #Curva Roc
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('Verdadeiro Positivo')
    plt.xlabel('Falso Positivo')
    plt.show()
```

```python
#knn
knn = KNeighborsClassifier(weights='distance', algorithm='auto', n_neighbors=15, n_jobs=4)

treina_modelo(knn,'KNN')
```

```python
rl = LogisticRegression(max_iter = 1000)
treina_modelo(rl,'Regressão Logística')
```

```python
dtc = DecisionTreeClassifier(criterion='gini', splitter='random', max_depth=25, min_samples_split=4,
                            min_samples_leaf=2)

treina_modelo(dtc, 'Arvore de Decisão')
```
